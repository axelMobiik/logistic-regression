<h1>Introduction to Logistic Regression</h1>
<p>
    We want to learn about Logistic Regression as a method fot <strong>Classification</strong>
    <br>Some examples of classification problems:
    <li>Spam versus "Ham" emails</li>
    <li>Loan Default (yes/no)</li>
    <li>Disease Diagnosis</li>
    Above were all examples of Binary Classification
</p>
<p>
    So far we've only seen regression problems where we try to predict a continuous value.
    <br>Although the name may be confusing at first, logistic regressions allows us to solve classification problems, where we are trying to predict discrete categories.
    <br>The convention for binary classification is to have two classes 0 and 1.
    <br>We can't use a normal linear regression model on binary groups. It lead to a good fit.
    <br>Instead we can transform our linear regression to a logistic regression curve.
    <br>The Sigmoid (aka Logistic) Function takes in any value and outputs it to be between 0 and 1 (Tettha(z) = 1/(1 + e^(-z)))
    <br>This means we can take our Linear Regression Solution and place it into the Sigmoid Function.
    <br>This results in a probability from 0 to 1 of belonging in the 1 class.
    <br>We use the logistic function to output a value ranging from 0 to 1. Based off of this probability we assign a class.
</p>
<p>
    After you train a logistic regression model on some training data, you will evaluate your model's performance on some test data.
    <br>You can use a confusion matrix to evaluate classification models.
    <br>We can use a confusion matrix to evaluate our model. For example, imagine testing for disease.
</p>

<p>
    Basic Terminology:
    <li>True Positive (TP)</li>
    <li>True Negative (TN)</li>
    <li>False Positive (FP)</li>
    <li>False Negative (FN)</li>
</p>
<p>
    Accuracy:
    <li>Overall, how often is it correct?</li>
    <li>(TP + TN) / total = 150/165 = 0.91</li>
</p>
<p>
    Misclassification Rate (Error Rate):
    <li>Overall, how often is it <strong>wrong</strong>?</li>
    <li>(FP + FN) / total = 15/165 = 0.09</li>
</p>
